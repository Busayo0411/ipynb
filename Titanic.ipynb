# Handle Missing Values.

### 1. Identify and fill missing values in the Age column with the mean.

### 2. Fill missing values in the Embarked column with the most frequent value(mode)

#### This is to figure out the mode of the embarked column.

```python
df['Embarked'].mode()[0]
```

#### This is to fill missing values on the embarked column with the mode.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df
```

## Dropping Rows and Columns:

### 3 Drop the Cabin column due to a high number of missing values.

#### This is to remove the cabin column.

```python
 df.drop(columns=['Cabin'])
```

### 4. Drop rows where fare is missing.

#### This is to drop/remove the rows where there are NaN values in the Fare column.

```python
df.dropna(subset=['Fare'])
```

## Changing Data Types:

### 5. Convert the Age column to an integer type.

#### This is to turn the Age column into a Interger type.

```python
df['Age'] = df['Age'].fillna(df['Age'].median()).astype('int')
```

```python
df
```

### 6. Convert the number data in the Survived column in the data to categorical data(Hint: Assign dead to 0,alive to 1).

```python
df['Survived'] = df['Survived'].map({0: 'dead', 1: 'alive'})
```

```python
df
```

## 1. How Many People Survived and How Many People Died?

```python
survived_count = df['Survived'].value_counts()
print("Number of people who survived:", survived_count['alive'])
print("Number of people who died:", survived_count['dead'])
```

## 2. How Many Passengers were in first class, second class and third class?

```python
# Count the number of passengers in each class
class_counts = df['Pclass'].value_counts()
# Display the counts
for pclass, count in class_counts.items():
    print(f"Number of passengers in class {pclass}:{count}")
```

## 3. Find the number of Male and Female passengers.

```python
sex_counts = df['Sex'].value_counts()
for sex,count in sex_counts.items():
    print(f"Number of {sex} passengers: {count}")
```

## 4. Who has Better Chance of Survival?

```python
# Grouping by 'Sex' and 'Pclass', and calculating and survival rate
survival_by_gender_class = df.groupby(['Sex', 'Pclass'])['Survived'].value_counts(normalize=True).unstack()
# Displaying the survival rate by gender and class
print("Survival Rate by Gender and Class:\n", survival_by_gender_class)
# Get the gender class with the highest survival rate
gender_class_with_highest_survival_rate = survival_by_gender_class['alive'].idxmax()
print("\ngender and class with the highest survival rate:", gender_class_with_highest_survival_rate)
```

```python
_# Calculate survival rate by passenger class
survival_rate_by_class = df.groupby(['Pclass'])['Survived'].value_counts(normalize=True).unstack()
# Displaying the survival rate by passenger class
print("Survival Rate by Passenger Class:")
print(survival_rate_by_class)
# Get the passenger class with the highest survival rate
class_with_highest_survival_rate = survival_rate_by_class['alive'].idxmax()
print("\npassenger class with the highest survival rate:", class_with_highest_survival_rate)
```
